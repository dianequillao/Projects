```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
import os
from string import ascii_lowercase
import re
import pandas as pd
import numpy as np
```

```python
#Define range of page urls
page_numbers = list((range(0, 5600, 100)))
print(page_numbers)
```

```python
def make_soup(url):
    thepage = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(thepage, "html.parser")
    return soupdata
```

```python
resorts=[]
for n in page_numbers: #for each 100 page numbers, use bs4 to get the html. table = the specific data table on the page    
    soup = make_soup("http://www.myresortnetwork.com/ForSale-stateS.asp?state=Hawaii+&order=Name&offset=" + str(n) + "")
    table = soup.find_all("table")[4]
    
    
    rows= table.find_all('tr')


    for row in rows:
        items=row.find_all('td')
        items = [item.text.strip() for item in items]
        resorts.append(items)

```

Creating new columns for the Minimum Price and Maximum Price ranges.

```python
def parse_min_range(l):
    try:
        return int(l[0].split('$')[1].strip())
    except:
        return 'NaN'
```

```python
new_df['PriceMin'] = new_df['PriceRange'].apply(parse_min_range)
```

```python
def parse_max_range(l):
    try:
        return (l[1].split('$')[1].strip())
    except:
        return 'NaN'
    
```

```python
new_df['PriceMax'] = new_df['PriceRange'].apply(parse_max_range)
```
